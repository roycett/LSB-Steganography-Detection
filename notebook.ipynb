```python
import tensorflow as tf
from tensorflow.keras import layers, models
import pandas as pd
import numpy as np
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
df1 = pd.read_csv('/kaggle/input/steganography-lsb-features-dataset/features_train_70000.csv')
df2 = pd.read_csv('/kaggle/input/steganography-lsb-features-dataset/features_test_70000.csv')

df1[' Complexity'] = pd.to_numeric(df1[' Complexity'], errors='coerce')
df1 = df1[pd.notna(df1[' Complexity'])]

df2[' Complexity'] = pd.to_numeric(df2[' Complexity'], errors='coerce')
df2 = df2[pd.notna(df2[' Complexity'])]

x_train = df1.iloc[:, 0:8]
y_train = df1.iloc[:, 8]
x_train = np.array(x_train).astype(np.float32)
y_train = np.array(y_train).astype(np.int32)

x_test = df2.iloc[:, 0:8]
y_test = df2.iloc[:, 8]
x_test = np.array(x_test).astype(np.float32)
y_test = np.array(y_test).astype(np.int32) 
```

```python
model = models.Sequential()
model.add(layers.Dense(32,activation='relu',input_shape=(8,)))
model.add(layers.Dense(32,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=30)
```

```python
from sklearn.metrics import accuracy_score

prob = model.predict(x_test)
y_pred = [1 if x >= 0.5 else 0 for x in prob]
acc = accuracy_score(y_test,y_pred)
print("Test accuracy:", acc)
```
